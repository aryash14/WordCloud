```python
import requests
from bs4 import BeautifulSoup
def get_video_links(url_link):
  #create response object
  r = requests.get(url_link)
  #create beautiful-soup object
  soup = BeautifulSoup(r.content)
  #find all links on web-page
  links = soup.findAll('a')
  #filter the link ending with .mp4
  video_links = [url_link + link['href'] for link in links if link['href'].endswith('mp4')]
 
  return video_links
url_link = "https://web.microsoftstream.com/video/a2fe1591-32c0-4d0f-832a-bc709a29b04a?list=studio"
video_links = get_video_links(url_link)
```

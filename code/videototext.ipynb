```python
audio = AudioSegment.from_wav("sample.wav")
#minimum time in ms to be considered silence, currently set to 1000ms
chunks = split_on_silence(audio, min_silence_len=1000, silence_thresh=-40)
```

```python
len(chunks)
```

```python
out = open("text1.txt", mode = "w+")
import os
try:
    os.mkdir("audio_chunks")
except(FileExistsError):
    pass
os.chdir("audio_chunks")
# for _, chunk in enumerate(chunks):0
for i in range(284, len(chunks)):
    # chunk = chunks[i]
    # Create 0.5 seconds silence chunk
    # chunk_silent = AudioSegment.silent(duration = 10)
    # add 0.5 sec silence to beginning and 
    # end of audio chunk. This is done so that
    # it doesn't seem abruptly sliced.
    # audio_chunk = chunk_silent + chunk + chunk_silent
    # audio_chunk.export("./chunk{}.wav".format(i), format = "wav")
    file_name = "chunk{}.wav".format(i)
    r = sr.Recognizer()
    with sr.AudioFile(file_name) as source:
        audio_f = r.listen(source)
    try:
        rec = r.recognize_google(audio_f)
        # write the output to the file.
        # print(rec)
        out.write(rec + ". ")
    except sr.UnknownValueError:
        pass
os.chdir('../')
out.close()
```
